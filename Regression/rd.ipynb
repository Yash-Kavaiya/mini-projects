## Theory
Resistors play a crucial role in electronics and circuits for their use in controlling electrical current. They are simply a small device that limits the flow of current throughout a circuit, or in other words, impedes the flow of current through resistance. Hence, the term resistor comes to mind when describing this type of electronic device.

A resistor takes advantage of the fact that it follows what we call Ohm's Law where a resistor placed under a potential V is proportional to current I and overall resistivity value of a resistor. To be more exact, Ohm's law writes
 $$V = I*R$$



where voltage is related to both current and resistor values. This means that under a given potential V and current I, we can determine the total resistivity of a resistor at any place in a circuit. The units for resistance is in terms of Ohms(), or otherwise

Now, suppose that we have an unknown resistor. To determine the value of a resistor we must rst provide a voltage across it along with a known value of current. The equipment that we use would include a power supply source that
does this for us. Once a power output is provided, we can measure the resistor by determining the value of current with a given voltage output. After measuring values for voltage output and current, we should expect a linear model as an
expectation for understanding the value of a resistor. $$Ohms = \frac{Volts}{Amperes}$$   
$$ 1 Ohm= \frac{1 Volts} {1 Ampere }$$

First of all, we import the necessary NumPy,Seaborn and Matplotlib libraries in our code

```python
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import matplotlib
```

Next, we load the data.

```python
data = np.loadtxt("resistance_data.txt")
x  =  data [ :,0]# separate  data  into  x  values
y  =  data [ :,1 ]# separate  data  into  y  values
n  =len( x )
```

First, we calculate the sum. Of course, we could use the for loop, but this method is inefficient. It is much better to use the NumPy library, which works with arrays, and all calculations are much faster. We will use special functions of the library.

```python
denominator = x.dot(x) - x.mean() * x.sum() 
```

Now you understand why it is so good to use the NumPy library. Now calculate the numerators of the coefficients a and b.

```python
a = ( x.dot(y) - y.mean()*x.sum() ) / denominator 
b = ( y.mean() * x.dot(x) - x.mean() * x.dot(y) ) / denominator
```

Now calculate our predicted Yhat

```python
Yhat = a*x + b
```

Plot a graph of experiment data and predicated value

```python
matplotlib.rcParams['font.size'] = 24
sns.set_style("darkgrid")
plt.figure(figsize=(14,8) )
plt.xlabel("Current(A)")
plt.ylabel("Voltage (V)" )
plt.title ("Current−Voltage Plot")
plt.scatter(x, y)
plt.plot(x, Yhat, ls='--', c='r')
plt.show() 
```

Here red line represent linear regression model and scatter plot are represent experiment value 

### Determination of the error of linear regression

```python
d1 = y - Yhat
d2 = y - y.mean()
r2 = d1.dot(d1) / d2.dot(d2)
accuracy= (1-r2)*100

print("Resistance value :- ",a)
print("Error in our linear regression:- ",r2*100," %")
print("Accurity of linear regression:- ",accuracy ,"%")
```
